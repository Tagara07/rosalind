[Complementing a Strand of DNA](https://rosalind.info/problems/revc/)

```python
with open(r'./data/q3_rosalind_revc.txt') as f:
    dna_str = f.read()
    f.close()
```

Solution 1

```python
from Bio.Seq import Seq
```

```python
my_str = Seq(dna_str)
my_str
```

```python
print(my_str.reverse_complement())
```

```python
print(len(dna_str))
print(len(my_str.reverse_complement()))
```

Solution 2

```python
match_dict = {'A':'T', 
              'C':'G',
              'G':'C',
              'T':'A'}
```

```python
reversed_str = []
for nuc in dna_str.replace('\n', ''):
    # print(nuc)
    reversed_str.append(match_dict[nuc])

reversed_str = ''.join(reversed_str[::-1])
print(reversed_str)
```
