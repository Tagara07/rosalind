```python
with open(r'./data/02_rosalind_bins.txt', 'r') as f:
    file = f.readlines()
    f.close()

n = int(file[0].replace('\n', ''))
m = int(file[1].replace('\n', ''))
str1 = file[2].replace('\n', '').split(' ')
list1 = [int(num) for num in str1]
str2 = file[3].replace('\n', '').split(' ')
list2 = [int(num) for num in str2]
print(f'n: {n}')
print(f'm: {m}')
print(f'list1: {list1}')
print(f'list2: {list2}')
```

Solution

```python
def binary_search(list, target):
    left = 0
    right = len(list) - 1

    while left <= right:
        mid = (left + right) // 2
        if list[mid] == target:
            return mid + 1
        elif list[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
result_list = []

for i in list2:
    result_list.append(binary_search(list1, i))
    
result_str = ' '.join([str(num) for num in result_list])

jf.gen_txt(result_str)
```
